```python
import scipy.stats as stats
import numpy as np
```

```python
z_min, z_max, step = -5, 5, 0.1  # Covering most of the normal distribution
z_values = np.arange(z_min, z_max + step, step)

# Compute probabilities for each chunk (area under curve for each interval)
cdf_values = stats.norm.cdf(z_values)
probabilities = np.diff(cdf_values)  # Differences between consecutive CDF values

total_probability = np.sum(probabilities)

# Print the results
print(f"Approximated total probability: {total_probability:.6f}")
```
